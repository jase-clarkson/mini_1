```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='serif')
from sklearn.decomposition import PCA
PATH = 'data'
df = pd.read_pickle('{}/top_500_returns.pkl'.format(PATH))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
pca = PCA()
pca.fit(df)
top = -1 * pca.components_[0]
axs[0].bar([i for i in range(len(top))], top, color='black')
axs[0].set_title('Eigenvector Loadings')
axs[1].set_title('Loadings Distribution')
axs[1].hist(top, bins=50, density=True, linewidth=1.2, edgecolor='black', color='white') 
fig.savefig('figs/top_ev.pdf', format='pdf', dpi=1200, bbox_inches='tight')
```
