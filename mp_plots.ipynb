```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 4))
def plot_mp(ax, p, n):
    gamma = p/n
    min, max = rmt.mp_eval_bounds(gamma)
    x = np.linspace(min, max, 5000)
    ax.plot(x, rmt.mp_pdf(x, p, n), c='k')
    ax.set_title(r'$\gamma={}$'.format(gamma))
    ax.set_xlabel('Eigenvalue')
    ax.set_ylabel('Probability Density')

plot_mp(ax1, 1, 4)
plot_mp(ax2, 2, 2)
plot_mp(ax3, 500, 250)
plt.tight_layout()
# plt.show()
fig.savefig('mp_plots.eps', format='eps', dpi=1200, bbox_inches='tight')
```
